```python
def add_cols_to_standings(df,col,new_col1,new_col2,new_col3,new_col4):    # will work because only need 'Pts' column
    new_col1 = col.diff().fillna(0)
    df.insert(10,"ptd",new_col1,True)

    new_col2 = [col[:4].diff().fillna(0).sum()]*len(df['Squad'])
    df.insert(11,'ptd_top_4',new_col2,True)

    new_col3 = [col[-5:].diff().fillna(0).sum()]*len(df['Squad'])
    df.insert(12,'ptd_bottom_5',new_col3,True)

    new_col4 = [col[-3:].sum()]*len(df['Squad'])
    df.insert(10,"pts_bottom_3",new_col4,True)
```

```python
# CREATING SERIES OF SUBSETS OF ALL TEAMS
def top_teams(apertura,clausura,ap_champ,ap_top_2,ap_top_3,ap_top_4,ap_top_5,ap_bottom_5,ap_bottom_3,cl_champ,cl_top_2,cl_top_3,cl_top_4,cl_top_5,cl_bottom_5,cl_bottom_3):
    ap_champ.extend(apertura.iloc[:1,1].copy())
    ap_top_2.extend(apertura.iloc[:2,1].copy())
    ap_top_3.extend(apertura.iloc[:3,1].copy())
    ap_top_4.extend(apertura.iloc[:4,1].copy())
    ap_top_5.extend(apertura.iloc[:5,1].copy())
    ap_bottom_5.extend(apertura.iloc[-5:,1].copy())
    ap_bottom_3.extend(apertura.iloc[-3:,1].copy())

    cl_champ.extend(clausura.iloc[:1,1].copy())
    cl_top_2.extend(clausura.iloc[:2,1].copy())
    cl_top_3.extend(clausura.iloc[:3,1].copy())
    cl_top_4.extend(clausura.iloc[:4,1].copy())
    cl_top_5.extend(clausura.iloc[:5,1].copy())
    cl_bottom_5.extend(clausura.iloc[-5:,1].copy())
    cl_bottom_3.extend(clausura.iloc[-3:,1].copy())
```

```python
# FUNCTION FOR CLEAINING FIXTURES TABLES

def clean_fixtures(apertura,clausura,ap_champ,ap_top_2,ap_top_3,ap_top_4,ap_top_5,ap_bottom_5,ap_bottom_3,cl_champ,cl_top_2,cl_top_3,cl_top_4,cl_top_5,cl_bottom_5,cl_bottom_3):

    # apertura

    #Losses by 2+ goals by not top 4, USE .all() WHEN DOING NOT EQUAL TO
    for index,row in apertura.iterrows():
        if ((row['Home']==pd.Series(ap_top_4)).any() & (row['Away']!=pd.Series(ap_top_4)).all()) and int(row['Score'][2])-int(row['Score'][0]) >=2:
            apertura.loc[index,'top_4_L_by2goals']=1  
        elif ((row['Away']==pd.Series(ap_top_4)).any() & (row['Home']!=pd.Series(ap_top_4)).all()) and int(row['Score'][2])-int(row['Score'][0]) <=-2:
            apertura.loc[index,'top_4_L_by2goals']=1


    #Losses against bottom 5
    for index,row in apertura.iterrows():
        if ((row['Home']==pd.Series(ap_top_4)).any() & (row['Away']==pd.Series(ap_bottom_5)).any()) and int(row['Score'][2])-int(row['Score'][0]) >=1:
            apertura.loc[index,'top_4_L_bybottom5']=1
        elif ((row['Away']==pd.Series(ap_top_4)).any() & (row['Home']==pd.Series(ap_bottom_5)).any()) and int(row['Score'][2])-int(row['Score'][0]) <=-1:
            apertura.loc[index,'top_4_L_bybottom5']=1



    #Champ Losses against top 6
    for index,row in apertura.iterrows():
        if ((row['Home']==pd.Series(ap_champ)).any() & (row['Away']==pd.Series(ap_top_5)).any()) and int(row['Score'][2])-int(row['Score'][0]) >=1:
            apertura.loc[index,'champ_L_bytop5']=1
        elif ((row['Away']==pd.Series(ap_champ)).any() & (row['Home']==pd.Series(ap_top_5)).any()) and int(row['Score'][2])-int(row['Score'][0]) <=-1:
            apertura.loc[index,'champ_L_bytop5']=1


    # clausura

    #Losses by 2+ goals by not top 4, USE .all() WHEN DOING NOT EQUAL TO
    for index,row in clausura.iterrows():
        if ((row['Home']==pd.Series(cl_top_4)).any() & (row['Away']!=pd.Series(cl_top_4)).all()) and int(row['Score'][2])-int(row['Score'][0]) >=2:
            clausura.loc[index,'top_4_L_by2goals']=1  
        elif ((row['Away']==pd.Series(cl_top_4)).any() & (row['Home']!=pd.Series(cl_top_4)).all()) and int(row['Score'][2])-int(row['Score'][0]) <=-2:
            clausura.loc[index,'top_4_L_by2goals']=1


    #Losses against bottom 5
    for index,row in clausura.iterrows():
        if ((row['Home']==pd.Series(cl_top_4)).any() & (row['Away']==pd.Series(cl_bottom_5)).any()) and int(row['Score'][2])-int(row['Score'][0]) >=1:
            clausura.loc[index,'top_4_L_bybottom5']=1
        elif ((row['Away']==pd.Series(cl_top_4)).any() & (row['Home']==pd.Series(cl_bottom_5)).any()) and int(row['Score'][2])-int(row['Score'][0]) <=-1:
            clausura.loc[index,'top_4_L_bybottom5']=1



    #Champ Losses against top 6
    for index,row in clausura.iterrows():
        if ((row['Home']==pd.Series(cl_champ)).any() & (row['Away']==pd.Series(cl_top_5)).any()) and int(row['Score'][2])-int(row['Score'][0]) >=1:
            clausura.loc[index,'champ_L_bytop5']=1
        elif ((row['Away']==pd.Series(cl_champ)).any() & (row['Home']==pd.Series(cl_top_5)).any()) and int(row['Score'][2])-int(row['Score'][0]) <=-1:
            clausura.loc[index,'champ_L_bytop5']=1
```

```python
def add_pk(performance,ap_standing,cl_standing):
    new_header = performance.iloc[0]
    performance = performance[1:]
    performance.columns = new_header
    performance = performance.iloc[:,11:13].astype(int).sort_values(['PKatt'],ascending=False)
    pk_top_3 = [performance['PKatt'][:3].astype(int).sum()]*len(ap_standing['Squad'])    #divided by 2 for the 2 different Conferences
    ap_standing.insert(14,'pk_top_3',pk_top_3,True)
    cl_standing.insert(14,'pk_top_3',pk_top_3,True)
```
